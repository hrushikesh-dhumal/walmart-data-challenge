</style><svg style="display: none;"><defs id="MathJax_SVG_glyphs"></defs></svg><h2>Introduction</h2>

<p>In order for customers to find our products while browsing the <a href="https://www.hackerrank.com/external_redirect?to=http://walmart.com/browse" target="_blank">walmart.com</a> site, we need to arrange the products onto virtual shelves, for example, "Smart TVs".</p>

<p>In this problem we ask for a machine learning solution to the task of assigning products to shelves.</p>

<p>The training data consists of detailed product information from 32 shelves. Your algorithm should suggest all shelves that each product should appear for the provided test set, which contains product information from the same 32 shelves. Each product can appear on more than one shelf.</p>

<h2>Dataset</h2>

<p>We provide a training dataset of 10593 products with various detailed product data, such as "Product Name", "Product Description", and various other attributes about the product. The columns are delimited by tabs and the data is found in <code>train.tsv</code>, with each row describing a different product.
These products are found on a total of 32 shelves. The shelves are labeled by integer values, and a list of all shelves that these products appear is given in the <code>tag</code> column for the training data.</p>

<p>We also provide a test set of 10593 products, without the labels, in <code>test.tsv</code>.
The goals is to predict the list of shelves that these products are assigned to.</p>

<p>You can download the zip file provided <a href="https://www.hackerrank.com/external_redirect?to=https://s3.amazonaws.com/hr-testcases-us-east-1/25053/assets/products-shelves-tagging-dataset.zip" target="_blank">here</a>. The zip file contains both the files <code>train.tsv</code> and <code>test.tsv</code>.</p>

<h2>Submission Details</h2>

<p>You are required to upload the following three files:</p>

<ul>
<li><p>The output file, <code>tags.tsv</code> (max allowed size is <code>10MB</code>). The file should contain each product from the file <code>test.tsv</code> followed by a list of the shelves for the product contained within square brackets.</p>

<p>A valid output file has the following format:</p>

<pre><code>item_id tag
10593   [4483]
10594   [581514]
10595   [4483]
10596   [4537]
10597   [1229817, 1229821]
</code></pre>

<p>Note that:</p>

<ul><li>The first line of the output file should contain the header, with <code>item_id</code>, and <code>tag</code> as the column names separated by a <code>tab</code>.</li>
<li>Products can appear on multiple shelves.</li>
<li>There are no limits on the number of shelves a product can be placed.</li></ul></li>
<li><p>A <em>PDF</em> file (maximum allowed size is <code>4MB</code>) providing the findings and justification on the following topics:</p>

<ul><li>Write a few lines about training dataset quality and any errors found in the training dataset.</li>
<li>Explain the data preprocessing steps.</li>
<li>Explain and justify the model you've chosen for the prediction.</li></ul></li>
<li><p>The source code of your approach for this task. Upload a <em>zip</em> file (maximum allowed size is <code>5MB</code>) with all relevant files to reproduce your results. The submitted file must have a <code>README</code> file with a detailed description about how to run the model to predict the shelves list for the products and generate the <code>tags.tsv</code>. Do not forget to include links to any external libraries or packages you use for the generation of your model.</p>

<p>There is no limit on execution time, but the code should generate the output file: <code>tags.tsv</code>.</p></li>
</ul>

```python
def get_distribution(pd_series):
    '''
    Function to get distribution in Pandas series
    Input: Pandas series
    Output: Pandas series
    '''
    
    distribution = {}
    for (idx, row) in pd_series.iteritems():
        val =str(row)
        if val in distribution:
            distribution[val] += 1
        else:
            distribution[val] = 1
                
    return ut.pd.Series(distribution)
```

## Pre processing
Perform following NLP task on data-
1. Lower case
2. Remove punctuation
3. Remove numbers
4. Remove stop words
5. Stem the words using PorterStemmer

```python
def remove_punctuation(string):
    '''
    Source: http://stackoverflow.com/questions/34293875/how-to-remove-punctuation-marks-from-a-string-in-python-3-x-using-translate
    '''
    # pass the translator to the string's translate method.
    return string.translate(TRANSLATOR_REMOVE_PUNCTUATION)
```

```python
def lowercase_check_for_stopword_stem(token):
    
    token = token.lower()
    if token not in NLTK_STOPWORDS:
        return True, STEMMER.stem(token)
    else:
        return False, None
```

```python
def remove_punctuation_lowercase_check_for_stopword_stem(string):
    
    string = remove_punctuation(string)
    
    # Not using nltk.word_tokenize(s) cause it takes long time to process.
    # Can be considered in case of complex text    
    token_list = string.split()

    op = []

    for token in token_list:
        a,b = lowercase_check_for_stopword_stem(token)
        if a:
            op.append(b)
  
    return ' '.join(op)
```
